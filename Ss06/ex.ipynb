```python
"""
Tạo một mảng 2D kích thước 3x3 với các giá trị ngẫu nhiên.
Tạo một mảng 1D có kích thước 3 và thực hiện phép cộng giữa mảng 2D và mảng 1D, áp dụng broadcasting.
Thực hiện phép trừ giữa mảng 2D và mảng có kích thước 1x3.
Kiểm tra kết quả và giải thích quá trình broadcasting.
Input
Output
 [[5 6 8]
 [7 2 4]
 [1 6 3]]
 [10 20 30]
Kết quả phép cộng A + B:
 [[15 26 38]
 [17 22 34]
 [11 26 33]]
Kết quả phép trừ A - B:
 [[ -5 -14 -22]
 [ -3 -18 -26]
 [ -9 -14 -27]]
"""
import numpy as np
array_1d = np.random.randint(1,101,(3))
array_2d = np.random.randint(1,101,(3,3))
print(array_1d)
print(array_2d)
print(f"Kết quả phép cộng:\n {np.add(array_1d,array_2d)}")
print(f"Kết quả phép trừ:\n {np.subtract(array_1d,array_2d)}")
```

```python
"""
Tạo một mảng 1D với các số ngẫu nhiên từ 1 đến 20.
Sử dụng vectơ hóa để tính bình phương và căn bậc hai của tất cả các phần tử trong mảng mà không cần vòng lặp.
Sử dụng hàm ufunc như np.add(), np.multiply(), np.sqrt() để thực hiện các phép toán giữa mảng và giá trị số.
Thực hiện phép tính với một mảng 2D và so sánh tốc độ với phương pháp vòng lặp thủ công.
"""
import numpy as np
import time
array_1d = np.random.randint(1,21,1)
print(f"Array_1d: {array_1d}")
print(f"Array_1d square: {array_1d**2}")
print(f"Array_1d square: {np.sqrt(array_1d)}")

print(f"Add 5: {np.add(array_1d, 5)}")
print(f"Multiply by 3: {np.multiply(array_1d, 3)}")
print(f"Sqrt (ufunc): {np.sqrt(array_1d)}")


array_2d = np.random.randint(1, 21, (500, 500))

start_vec = time.time()
result_vec = array_2d * 2 + 5
end_vec = time.time()

start_loop = time.time()
rows, cols = array_2d.shape
result_loop = np.zeros_like(array_2d)

for i in range(rows):
    for j in range(cols):
        result_loop[i, j] = array_2d[i, j] * 2 + 5

end_loop = time.time()

print("\n--- Speed Comparison ---")
print(f"Vectorized time: {end_vec - start_vec:.6f} sec")
print(f"Loop time:        {end_loop - start_loop:.6f} sec")
```

```python
"""
Tạo một mảng 2D kích thước 5x5 với các giá trị ngẫu nhiên.
Tạo một mảng 1D kích thước 5 và thực hiện các phép toán với broadcasting (cộng, trừ) giữa mảng 2D và mảng 1D.
Áp dụng các hàm ufunc như np.sqrt(), np.add(),.. vào mảng 2D.
Đo tốc độ thực thi của việc sử dụng broadcasting và ufunc và so sánh với vòng lặp truyền thống.
Input
Output
 [[ 8 12 14  6  4]
 [18 13  2  2 12]
 [ 6 18  5 10 10]
 [ 9  2  4 12 12]
 [ 2  3  2 17  3]]
 [10 20 30 40 50]
Kết quả cộng (broadcasting):
 [[18 32 44 46 54]
 [28 33 32 42 62]
 [16 38 35 50 60]
 [19 22 34 52 62]
 [12 23 32 57 53]]
Kết quả trừ (broadcasting):
 [[ -2  -8 -16 -34 -46]
 [  8  -7 -28 -38 -38]
 [ -4  -2 -25 -30 -40]
 [ -1 -18 -26 -28 -38]
 [ -8 -17 -28 -23 -47]]
Kết quả căn bậc hai (ufunc):
 [[2.82842712 3.46410162 3.74165739 2.44948974 2.        ]
 [4.24264069 3.60555128 1.41421356 1.41421356 3.46410162]
 [2.44948974 4.24264069 2.23606798 3.16227766 3.16227766]
 [3.        1.41421356 2.        3.46410162 3.46410162]
 [1.41421356 1.73205081 1.41421356 4.12310563 1.73205081]]
Kết quả cộng 5 vào mảng (ufunc):
 [[13 17 19 11  9]
 [23 18  7  7 17]
 [11 23 10 15 15]
 [14  7  9 17 17]
 [ 7  8  7 22  8]]
Thời gian thực thi vòng lặp thủ công: 0.001234 giây
"""
import numpy as np
import time

# -------------------------------
# 1. Tạo mảng 2D và 1D
# -------------------------------
arr_2d = np.random.randint(1, 21, (5, 5))
arr_1d = np.random.randint(10, 60, 5)

print(arr_2d)
print(arr_1d)

# -------------------------------
# 2. Broadcasting (cộng, trừ)
# -------------------------------
add_result = arr_2d + arr_1d
sub_result = arr_2d - arr_1d

print("Kết quả cộng (broadcasting):")
print(add_result)

print("Kết quả trừ (broadcasting):")
print(sub_result)

# -------------------------------
# 3. Ufunc: sqrt, add
# -------------------------------
sqrt_result = np.sqrt(arr_2d)
add5_result = np.add(arr_2d, 5)

print("Kết quả căn bậc hai (ufunc):")
print(sqrt_result)

print("Kết quả cộng 5 vào mảng (ufunc):")
print(add5_result)

# -------------------------------
# 4. So sánh tốc độ: broadcasting vs loop
# -------------------------------

# Broadcasting
start_bc = time.time()
bc_result = arr_2d + arr_1d
end_bc = time.time()
bc_time = end_bc - start_bc

# Vòng lặp thủ công
start_loop = time.time()
loop_result = np.zeros_like(arr_2d)

for i in range(arr_2d.shape[0]):
    for j in range(arr_2d.shape[1]):
        loop_result[i, j] = arr_2d[i, j] + arr_1d[j]

end_loop = time.time()
loop_time = end_loop - start_loop

print(f"Thời gian thực thi broadcasting: {bc_time:.6f} giây")
print(f"Thời gian thực thi vòng lặp thủ công: {loop_time:.6f} giây")
```

```python
"""
Tạo một ma trận vuông 3x3 với các giá trị ngẫu nhiên.
Tính ma trận nghịch đảo của ma trận vừa tạo.
Tính giá trị riêng và véc tơ riêng của ma trận.
So sánh kết quả giữa ma trận gốc và ma trận nghịch đảo.
Input
Output
[[6 9 7]
 [2 1 5]
 [7 8 3]]
Ma trận nghịch đảo:
 [[ 0.52941176 -0.23529412 -0.29411765]
 [-0.29411765  0.41176471 -0.05882353]
 [-0.05882353 -0.17647059  0.29411765]]
Giá trị riêng của ma trận: [15.67207479  0.27335687  -2.94543166]
Véc tơ riêng của ma trận:
 [[ 0.58528917 -0.3277603  -0.50732504]
 [ 0.72147434 -0.26197579  0.76096577]
 [ 0.36829758  0.90447319 -0.40544259]]
"""
import numpy as np

# 1. Tạo ma trận vuông 3x3 với giá trị ngẫu nhiên 1→10
A = np.random.randint(1, 11, (3, 3))
print(A)

# 2. Tính ma trận nghịch đảo
A_inv = np.linalg.inv(A)
print("Ma trận nghịch đảo:")
print(A_inv)

# 3. Tính giá trị riêng và véc tơ riêng
eigen_values, eigen_vectors = np.linalg.eig(A)

print("Giá trị riêng của ma trận:")
print(eigen_values)

print("Véc tơ riêng của ma trận:")
print(eigen_vectors)
```
