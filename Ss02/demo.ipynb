```python
# Bài 5
def is_palindrome(n):
    str_n = str(n)
    return str_n == str_n[::-1]

n = int(input("Nhap vao 1 so nguyen duong : "))
if(n <= 0):
    print("Vui lòng nhập số nguyên dương!")
else:
    if is_palindrome(n):
        print(f"{n} là số đối xứng (palindrome).")
    else:
        print(f"{n} không phải là số đối xứng (palindrome).")
```

```python
# Bài 6
def factorial(n):
    result = 1
    for i in range(2,n+1):
        result *= i
    return result

n = int(input("Nhập một số nguyên dương: "))
if n < 0:
    print("Vui lòng nhập số nguyên dương!")
else:
    print(f"{n}! = {factorial(n)}")
```

```python
# Bài 7
def is_prime(n):
    if(n < 2):
        return False
    for i in range(2,int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

a = int(input("Nhập một số nguyên a: "))
b = int(input("Nhập một số nguyên b: "))
print(f"Các số nguyên tố trong khoảng [{a}, {b}]:")
for i in range(a,b+1) : 
    if is_prime(i):
        print(i,end=" ")
```
