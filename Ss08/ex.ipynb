```python
# Bài 5
import pandas as pd

# ============================
# 1) Tạo file students.csv
# ============================
students_data = """id,name
1,An
2,Binh
3,Chau
"""

with open("students.csv", "w", encoding="utf-8") as f:
    f.write(students_data)

# ============================
# 2) Tạo file scores.csv
# ============================
scores_data = """id,score
1,8
2,9
3,7
"""

with open("scores.csv", "w", encoding="utf-8") as f:
    f.write(scores_data)

# ============================
# 3) Đọc file bằng pandas
# ============================
students_df = pd.read_csv("students.csv")
scores_df = pd.read_csv("scores.csv")

# ============================
# 4) Merge theo id
# ============================
merged = pd.merge(students_df, scores_df, on="id")

# ============================
# 5) Tính min – max – mean cho từng sinh viên
# (Trường hợp thực tế mỗi sinh viên có nhiều điểm)
# ============================
report = merged.groupby(["id", "name"]).agg(
    score_min=("score", "min"),
    score_max=("score", "max"),
    score_mean=("score", "mean")
).reset_index()

print(report)

# ============================
# 6) Lưu vào report.csv
# ============================
report.to_csv("report.csv", index=False)
```

```python
# Bài 6
import numpy as np

# Tạo mảng a kích thước (4,1)
a = np.array([[1], [2], [3], [4]])
print("a =\n", a)

# Tạo mảng b kích thước (1,5)
b = np.array([[10, 20, 30, 40, 50]])
print("\nb =\n", b)

# Broadcasting tính a * b
result = a * b
print("\nKết quả a * b =\n", result)

# Áp dụng sqrt() và log()
sqrt_result = np.sqrt(result)
log_result = np.log(result)

print("\nsqrt(result) =\n", sqrt_result)
print("\nlog(result) =\n", log_result)

# Tính min, max, mean, std theo axis=1
min_vals = np.min(result, axis=1)
max_vals = np.max(result, axis=1)
mean_vals = np.mean(result, axis=1)
std_vals = np.std(result, axis=1)

print("\nMin theo axis=1 =", min_vals)
print("Max theo axis=1 =", max_vals)
print("Mean theo axis=1 =", mean_vals)
print("Std theo axis=1 =", std_vals)
```

```python
# Bài 7
import pandas as pd
import numpy as np

# ============================
# 1) Tạo DataFrame
# ============================
df = pd.DataFrame({
    "name": ["An", "Bình", "Châu", "Dũng"],
    "age": [20, np.nan, 22, 21],
    "score": [8.5, np.nan, 7.0, 9.0],
    "city": ["Hà Nội", "HCM", np.nan, "Hà Nội"]
})

print("Data gốc:")
print(df)

# ============================
# 2) Kiểm tra dữ liệu thiếu
# ============================
print("\nDữ liệu thiếu:")
print(df.isna())

# ============================
# 3) Điền dữ liệu thiếu
# ============================
df["age"] = df["age"].fillna(df["age"].mean())      # mean
df["score"] = df["score"].fillna(df["score"].median())  # median
df["city"] = df["city"].fillna("Unknown")           # Unknown

print("\nage_cleaned =", df["age"].tolist())
print("score_cleaned =", df["score"].tolist())
print("city_cleaned =", df["city"].tolist())

# ============================
# 4) Chuyển city → category
# ============================
df["city"] = df["city"].astype("category")

print("\ncity dtype =", df["city"].dtype)

# ============================
# 5) Tạo cột score_level bằng apply()
# ============================
def level(score):
    if score >= 8:
        return "Good"
    elif score >= 7:
        return "Average"
    return "Bad"

df["score_level"] = df["score"].apply(level)
print("\nscore_level =", df["score_level"].tolist())

# ============================

df["age_normalized"] = (df["age"] - df["age"].mean()) / df["age"].std()

print("\nage_normalized =", df["age_normalized"].tolist())
```

```python
# Bài 8
from pathlib import Path
from datetime import datetime
import json
import pandas as pd

# ==============================
# 1) Lấy toàn bộ file JSON
# ==============================
log_folder = Path("logs/")
json_files = list(log_folder.glob("*.json"))

records = []

# ==============================
# 2) Đọc từng file + lấy ngày từ tên
# ==============================
for file in json_files:
    # Ví dụ tên file: log_2025-05-21.json
    date_str = file.stem.split("_")[-1]  
    log_date = datetime.strptime(date_str, "%Y-%m-%d").date()

    # Đọc JSON
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)

    # Giả sử JSON có dạng:
    # {"user": "...", "action": "...", "duration": 12.5}
    record = {
        "date": log_date,
        "user": data.get("user"),
        "action": data.get("action"),
        "duration": data.get("duration", 0)
    }
    records.append(record)

# ==============================
# 3) Tạo DataFrame
# ==============================
df = pd.DataFrame(records)
print("DataFrame tổng hợp:")
print(df)

# ==============================
# 4) Tổng duration theo ngày
# ==============================
daily_report = df.groupby("date")["duration"].sum().reset_index()
print("\nBáo cáo theo ngày:")
print(daily_report)

# ==============================
# 5) Xuất ra CSV
# ==============================
daily_report.to_csv("daily_report.csv", index=False)
print("\nĐã tạo file daily_report.csv")
```
