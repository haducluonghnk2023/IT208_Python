```python
student = {
    "name":"Lan",
    "score":{
        "math":9,
        "english":8,
        "history":7 
    }
}
print(f"{student['score']["english"]}")
scores = student["score"].values()
average = sum(scores) / len(scores)
print(f"Điểm trung bình : {average}")
```

```python
products = {"A":120,"B":80,"C":300,"D":50}
products = {k : v for k , v in products.items() if v >= 100}
print(f"{products}")
products["E"] = 200
print(f"{products}")
print(f"{len(products)}")
```

```python
products = {"A": 120, "B": 80, "C": 300, "D": 50}

if "A" in products:
    print(f"Giá sản phẩm A: {products['A']}")
else:
    print("Sản phẩm A đã bị xóa")

sorted_products = dict(sorted(products.items(), key=lambda item: item[1]))
print("Danh sách sản phẩm sau khi sắp xếp theo giá:", sorted_products)

total_value = sum(products.values())
print("Tổng giá trị các sản phẩm:", total_value)

products = {k: v for k, v in products.items() if v <= 200}
print("Danh sách sau khi xóa sản phẩm giá > 200:", products)
```
