```python
# Bài 1
def subtract(a,b):
    return a - b

def multiply(a,b):
    return a * b
```

```python
# Bài 2
def is_prime(n):
    if n < 2 :
        return False
    for i in range(2,int(n**0.5)+1):
        if n % 2 == 0 :
            return False
    return True
```

```python
# Bài 3
def factorial(n):
    result = 1
    for i in range(2,n + 1):
        result *= i
    return result
```

```python
# Bài 4
def circle_area(radius):
    return 3.14159 * radius*radius
```

```python
# Bài 5
def is_leap_year(year):
    if year % 400 == 0:
        return True
    if year % 100 == 0:
        return False
    if year % 4 == 0:
        return True
    return False
```

```python
# Bài 6
def sum_of_digits(n):
    n = abs(n)
    total = 0
    while(n > 0):
        total += n % 10
        n //= 10
    return total
```

```python
#Bài 7
def handleReverseTheNumber(n):
    negative = n < 0
    n = abs(n)
    reversed_number = int(str(n)[::-1])
    return -reversed_number if negative else reversed_number
```

```python
# Bài 8 
def is_perfect_number(n):
    if n <= 1:
        return False
    total = 0
    for i in range(1,n):
        if n % i == 0:
            total += i
    return total == n
```
