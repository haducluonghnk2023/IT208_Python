```python
# Bài 2 
"""
Viết chương trình Python sử dụng csv.reader và csv.writer để:
Đọc dữ liệu từ file CSV có tên students.csv. Dữ liệu trong file có định dạng: name, age, grade
Sau khi đọc xong, in ra danh sách tất cả học sinh có grade >= 7.
Ghi dữ liệu này vào một file CSV mới có tên high_grade_students.csv.
Yêu cầu:
Sử dụng csv.reader để đọc dữ liệu từ file.
Dùng csv.writer để ghi lại dữ liệu vào file mới.
"""
import csv

high_students = []

with open("students.csv", mode="r", encoding="utf-8") as file:
    reader = csv.reader(file)
    next(reader)
    for row in reader:
        name, age, grade = row[0], row[1], float(row[2])

        if grade >= 7:
            print(row)
            high_students.append(row)

with open("high_grade_students.csv", mode="w", encoding="utf-8", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(high_students)

print("Đã ghi dữ liệu vào high_grade_students.csv")
```

```python
"""
Bài 3 
Viết chương trình Python thực hiện các công việc sau:
Mở một file JSON có tên user_data.json và đọc nội dung từ đó.
Tìm kiếm trong file JSON để in ra tên và tuổi của người có id = 3.
Thêm một đối tượng mới vào file JSON với thông tin của người dùng mới (id, name, age) và ghi lại vào file.
Yêu cầu:
Sử dụng json.load() và json.dump() để thao tác với file JSON.
Đảm bảo rằng dữ liệu được ghi lại đúng cách sau khi thêm thông tin mới.
"""
import json
with open("user_data.json",mode="r",encoding="utf-8") as file:
    data = json.load(file)
    
for i in data:
    if i["id"] == 3 :
        print(f"Tên: {i['name']} , Tuổi: {i['age']}")
        break

new_user = {
    "id": 4,
    "name": "Nguyen Van A",
    "age": 25
}
data.append(new_user)

with open("user_data.json", "w", encoding="utf-8") as file:
    json.dump(data, file, indent=4, ensure_ascii=False)

print("Đã thêm người dùng mới!")
```

```python
"""
Bài 6 :
Viết chương trình Python để:
Đọc dữ liệu từ file JSON inventory.json, chứa danh sách các sản phẩm với thông tin: product_id, name, quantity, price.
Thêm một sản phẩm mới vào file JSON.
Ghi lại dữ liệu JSON sau khi thêm sản phẩm mới.
Yêu cầu:
Sử dụng json.load() và json.dump() để thao tác với file JSON.
Đảm bảo rằng sản phẩm mới được thêm vào đúng cách.
"""
import json

with open("inventory.json", mode="r", encoding="utf-8") as file:
    data = json.load(file)

for i in data:
    print(f"Mã sản phẩm: {i['product_id']}, Tên sản phẩm: {i['name']}, Số lượng: {i['quantity']}, Giá: {i['price']}")

new_product = {
    "product_id": 2,
    "name": "Quần bò nam",
    "quantity": 5,
    "price": 500000
}
data.append(new_product)

with open("inventory.json", mode="w", encoding="utf-8") as file:
    json.dump(data, file, indent=4, ensure_ascii=False)

print("Đã thêm sản phẩm mới!")
```

```python
"""
Bài 8
Viết chương trình Python để:
Mở file log.txt, trong đó mỗi dòng chứa một timestamp và một thông điệp.
Đọc tất cả các dòng và in ra các dòng có timestamp trong ngày hôm nay.
Yêu cầu:
Sử dụng datetime để so sánh ngày tháng trong timestamp.
"""
from datetime import datetime
from pathlib import Path

today = datetime.today().date()
log_file = Path("log.txt")

with log_file.open("r", encoding="utf-8") as file:
    for line in file:
        parts = line.strip().split(" - ", 1)
        if len(parts) != 2:
            continue 
        
        timestamp_str, message = parts

        try:
            timestamp = datetime.strptime(timestamp_str, "%Y-%m-%d %H:%M:%S")
        except ValueError:
            continue

        if timestamp.date() == today:
            print(line.strip())
```
